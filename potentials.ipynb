```python
import matplotlib.pyplot as plt
import numpy as np
```

## Hyperbolic

```python
l = 6
kx = np.linspace(-3, 3, 1000)
V_q = -(l * (l + 1)) * (1 / np.cosh(kx))**2 + l**2
V_q *= 0.5
V_cl = -(l ** 2) * (1 / np.cosh(kx))**2 + (l ** 2)
V_cl *= 0.5
energies = [-0.5 * ((l-n)**2-l**2) for n in range(0, l)]
plt.figure(figsize=(8, 6))

plt.plot(
    kx, V_q, color='purple', lw=2,
    label=f'$\mathrm{{V_q(x)}}$ Quantum Potential'
)

plt.plot(
    kx, V_cl, color='royalblue', lw=2,
    label=f'$\mathrm{{V_{{cl}}(x)}}$ Classical Potential'
)

plt.fill_between(kx, V_q, V_cl, color='red', alpha=0.1)

for n, E in enumerate(energies):
    bound = abs(kx[abs(V_q - E).argmin()])
    plt.hlines(E, -bound, bound, colors='black', linestyles='--', lw=1.5)
    plt.text(-0.05, E + 0.5, fr'$\mathrm{{E_{{{n}}}}}$', color='black', va='center', fontsize=10)
plt.xlabel(r'Position $kx$', fontsize=12)
plt.ylabel(r'Energy E $\frac{{k^2\hbar^2}}{{m}}$', fontsize=12)
plt.title(r'Hyperbolic Pöschl–Teller Potential Well and Energy Levels', fontsize=14)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.xlim(-3.5, 3.5)
plt.tight_layout()
plt.show()
```

```python
from matplotlib.cm import plasma

l = 6
kx = np.linspace(-3, 3, 1000)

colors = [plasma(n / l) for n in range(l)]

plt.figure(figsize=(9, 6))

for n in range(l):
    V_q = -((l - n) * (l - n + 1)) * (1 / np.cosh(kx))**2 + (l - n)**2
    V_q *= 0.5
    
    plt.plot(
        kx, V_q,
        color=colors[n],
        lw=2,
        alpha=0.9 - 0.1 * n,
        label=fr'$\mathrm{{V_q^{{({n})}}(x)}}$'
    )

plt.title(r'Hyperbolic Pöschl–Teller Potentials', fontsize=14)
plt.xlabel(r'Position $kx$', fontsize=12)
plt.ylabel(r'Energy $E\,\frac{k^2\hbar^2}{m}$', fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=9, loc='upper left', frameon=False)
plt.xlim(-3.5, 3.5)
plt.tight_layout()
plt.show()
```

```python
l = 6
tol = 0.4
lower, upper = -np.pi/2+tol, np.pi/2-tol

kx = np.linspace(lower, upper, 1000)
V_q = (l * (l - 1)) * (1 / np.cos(kx))**2 - l**2
V_q *= 0.5
V_cl = (l ** 2) * (1 / np.cos(kx))**2 - l**2
V_cl *= 0.5
energies = [0.5 * ((l+n)**2-l**2) for n in range(0, l+1)]
plt.figure(figsize=(8, 6))

plt.plot(
    kx, V_q, color='purple', lw=2,
    label=f'$\mathrm{{V_q(x)}}$ Quantum Potential'
)

plt.plot(
    kx, V_cl, color='royalblue', lw=2,
    label=f'$\mathrm{{V_{{cl}}(x)}}$ Classical Potential'
)

plt.fill_between(kx, V_q, V_cl, color='red', alpha=0.1)

for n, E in enumerate(energies[:-1]):
    bound = abs(kx[abs(V_q - E).argmin()])
    plt.hlines(E, -bound, bound, colors='black', linestyles='--', lw=1.5)
    plt.text(-0.05, E + 1, fr'$\mathrm{{E_{{{n}}}}}$', color='black', va='center', fontsize=10)
plt.xlabel(r'Position $kx$', fontsize=12)
plt.ylabel(r'Energy E $\frac{{k^2\hbar^2}}{{m}}$', fontsize=12)
plt.title(r'Trigonometric Pöschl–Teller Potential Well and Energy Levels', fontsize=14)
plt.legend(fontsize=11)
plt.grid(alpha=0.3)
plt.xlim(lower, upper)
plt.ylim(min(V_cl.min(), V_q.min()) - l, max(energies))
plt.tight_layout()
plt.show()
```
