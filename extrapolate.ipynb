```python
t_eval = results[domains[0]]['t']
coeffs = {}
for d in domains:
    unbound = results[d]['unbound_x']
    hit_boundary = np.where(np.array(results[d]['unbound_edge_pct_-1']) > 0.1)
    if len(hit_boundary[0]) > 0:
        hit_boundary = hit_boundary[0][0]
    else:
        hit_boundary = len(unbound)
    
    plt.plot(t_eval[:hit_boundary], unbound[:hit_boundary], label=f"Domain {d}")
    # do a linear regression of t_eval[:hit_boundary] and unbound[:hit_boundary]
    c = np.polyfit(t_eval[:hit_boundary], unbound[:hit_boundary], 1)
    coeffs[d] = c
    poly = np.poly1d(c)
    plt.plot(t_eval[:hit_boundary], poly(t_eval[:hit_boundary]), linestyle='--', label=f"Fit {d}")
    

plt.xlabel("Time")
plt.ylabel("Unbound Expectation of x")
plt.title("Unbound Expectation of x vs Time for Increasing Domain Sizes")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# plt.savefig('unbound_x_vs_time.png')
plt.show()
```
